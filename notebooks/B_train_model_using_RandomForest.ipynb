```python
import pandas as pd
```

```python
import zipfile
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix 
```

```python
zip_path=r"C:\Users\DELL\Documents\GitHub\ProSpy\data\raw\archive (2).zip"
```

```python
with zipfile.ZipFile(zip_path, 'r') as z:
    train_df = pd.read_csv(z.open('train.csv'))
    test_df = pd.read_csv(z.open('test.csv'))
```

```python
X_train = train_df.drop('fake', axis=1)
y_train = train_df['fake']


```

```python
X_test = test_df.drop('fake', axis=1)
y_test = test_df['fake']
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
