```python
import pandas as pd
import zipfile
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
zip_path=r"C:\Users\DELL\Documents\GitHub\ProSpy\data\raw\archive (2).zip"
with zipfile.ZipFile(zip_path, 'r') as z:
    train_df = pd.read_csv(z.open('train.csv'))
    test_df = pd.read_csv(z.open('test.csv'))
```

```python
X_train = train_df.drop('fake', axis=1)
y_train = train_df['fake']

X_test = test_df.drop('fake', axis=1)
y_test = test_df['fake']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy : ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
